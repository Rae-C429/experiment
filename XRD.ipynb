```python
import numpy as np
```

```python
def xRD(meterial ,hkl, mno, pqr, a, b, c):
    print(meterial,'(',hkl[0],hkl[1],hkl[2],')')
    lam = 1.5418
    Ka = 2 * np.pi / a
    Kb = 2 * np.pi / b
    Kc = 2 * np.pi / c
    hkl_K = np.array([hkl[0]*Ka, hkl[1]*Kb, hkl[2]*Kc])
    # print(hkl_K)
    hkl_kval =  np.sqrt(hkl_K[0]**2 + hkl_K[1] **2 + hkl_K[2] **2)
    th = np.arcsin(lam * hkl_kval /(np.pi * 4)) * 180 / np.pi
    print('tth:',2*th)
    mno_K = np.array([mno[0]*Ka, mno[1]*Kb, mno[2]*Kc])
    mno_kval =  np.sqrt(mno_K[0]**2 + mno_K[1] **2 + mno_K[2] **2)
    dot = hkl_K[0]*mno_K[0]+hkl_K[1]*mno_K[1]+hkl_K[2]*mno_K[2]

    ormaga1 = th + np.arccos(dot/(hkl_kval*mno_kval))* 180/ np.pi 
    ormaga2 = th - np.arccos(dot/(hkl_kval*mno_kval))* 180/ np.pi
    print ('W+:',ormaga1)
    print ('W-:',ormaga2)
    return hkl_kval
```

```python
def similar_axis(ycmo_K_val ,sto_meterial,sto_mno, sto_pqr, sto_a, sto_b, sto_c):
    Ka = 2 * np.pi / sto_a
    Kb = 2 * np.pi / sto_b
    Kc = 2 * np.pi / sto_c
    for A in range(1,5):
        for B in range(1,5):
            temp_K = np.array([0,0,0])
            # print(sto_mno, sto_pqr)
            for i in range(0,3):
                # print('!')
                temp_K[i] = A * sto_mno[i] + B * sto_pqr[i]
            # print(temp_K)
            K = [temp_K[0] * Ka, temp_K[1] * Kb, temp_K[2] * Kc]
            # print(K)
            K_val =  np.sqrt(K[0]**2 + K[1]**2 + K[2]**2)
            # print(K_val)
            # print(ycmo_K_val, type(ycmo_K_val))
            if ycmo_K_val - 1 < K_val and K_val < ycmo_K_val + 1:
                # print('!')
                sto_hkl = temp_K
                xRD(sto_meterial ,sto_hkl, sto_mno, sto_pqr, sto_a, sto_b, sto_c)
                
            # print("STO K_val:",sto_K_val)
    
```
