```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def opentxt(dataname):
    path = dataname
    with open(path, mode = 'r') as f:
        list = f.read().split('\n')
        list = list[0: -1]
    data_total = []
    for line in list:
        data = [float(num) for num in line.split('\t')]
        data_total.append(data)
    datas = np.array(data_total)    
    return datas
```

New quartz

glass signal

```python
'''
glass
'''
# XRD
# read file
da = opentxt('GLASS.xy')
Dx1 = da[:, 0]
Dy1 = np.log(da[:, 1])
# plot
fig, D1 = plt.subplots(figsize = (10, 5))
D1.plot(Dx1, Dy1, label = "glass")
D1.set_title("XRD")
D1.set_xlabel("tth""($\\theta$)")
D1.set_ylabel("intensity")
plt.legend()
plt.show()
```
