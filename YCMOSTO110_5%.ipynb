```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def opentxt(dataname):
    path = dataname
    with open(path, mode = 'r') as f:
        list = f.read().split('\n')
        list = list[0: -1]
    data_total = []
    for line in list:
        data = [float(num) for num in line.split('\t')]
        data_total.append(data)
    datas = np.array(data_total)    
    return datas
```
